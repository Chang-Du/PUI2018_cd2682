```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np
```

```python
np.random.seed()
ReprRand = np.random.randn(2,100)
print(ReprRand)
#100个随机点
```

```python
ReprRandAll = np.zeros((50, 2, 100))
for i in range(50):
    np.random.seed()
    ReprRandAll[i]=  ReprRand
print(ReprRandAll)
#50组随机点
```

```python
import pylab as pl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
%pylab inline
for i in range (50):
    plt.scatter(ReprRandAll[i][0], ReprRandAll[i][1], marker = 'x',color = 'blue')
     
plt.xlabel('x')
plt.ylabel('y')
plt.title('Distribution of Random Numbers')
plt.show() 
```
